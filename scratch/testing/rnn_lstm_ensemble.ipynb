```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.models import load_model

# Import the ModulationLSTMClassifier
from ml_wireless_classification.rnn_lstm_w_SNR_ensemble import ModulationLSTMClassifier

data_path = "../RML2016.10a_dict.pkl"

# Import the ModulationLSTMClassifier
from ml_wireless_classification.rnn_lstm_w_SNR_ensemble import ModulationLSTMClassifier

# Instantiate the classifier
modulation_classifier = ModulationLSTMClassifier(data_path=data_path)

model_name = "rnn_lstm_w_SNR_5_2_1_ensemble"
# Load the pre-trained RNN/LSTM model
model_path = f"../src/ml_wireless_classification/models/{model_name}.keras"
modulation_classifier.model = load_model(model_path)

# Load the data
with open(data_path, "rb") as f:
    data = pickle.load(f, encoding="latin1")
    modulation_classifier.data = data

# Call prepare_data to get the train/test splits and label encoder
train_data, test_data = modulation_classifier.prepare_data()
X_train, X_combined_train, y_train = train_data
X_test, X_combined_test, y_test = test_data
label_encoder = modulation_classifier.label_encoder
```
