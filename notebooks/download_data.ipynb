```python
import kagglehub
import os
import shutil

# Download the dataset
path = kagglehub.dataset_download("pinxau1000/radioml2018")
print("Path to dataset files:", path)

# Define target directory inside your repo
target_dir = os.path.join(os.getcwd(), "..", "data", "RML2018")
os.makedirs(target_dir, exist_ok=True)

# Move any .h5, .pkl, or .mat files found in the Kaggle download
for root, _, files in os.walk(path):
    for f in files:
        if f.endswith((".h5", ".pkl", ".mat", ".tar.gz")):
            src = os.path.join(root, f)
            dst = os.path.join(target_dir, f)
            print(f"Moving {src} → {dst}")
            shutil.move(src, dst)

print("✅ All dataset files moved to:", target_dir)
```
